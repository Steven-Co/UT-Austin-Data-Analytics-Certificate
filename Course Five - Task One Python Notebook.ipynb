```python
# Data types section
varOne = 25
varTwo = 25.0
varThree = varOne + varTwo
print(varThree)
```

```python
varOne = 25
varTwo = 'Hello'
# This will not work as the datatypes are incompatible
varThree = varOne + varTwo
print(varThree)
```

The error above is the result of trying to combine two incompatible datatypes. One potential solution is to convert both variables to strings to get the following value

```python
varThree = str(varOne) + varTwo
print(varThree)
```
